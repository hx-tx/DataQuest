# Introduction
This dataset looks at the Euro daily exchange rates between 1999 - 2021. The Euro is the official currency in most of the countries of the EU.

```python
import pandas as pd
exchange_rates = pd.read_csv("euro-daily-hist_1999_2020.csv")
```

# Data Cleaning

Our focus in the project will be on the exchange rate between the euro and the American dollar. 

```python
exchange_rates.rename(columns={'[US dollar ]': 'US_dollar',
                               'Period\\Unit:': 'Time'},
                      inplace=True)
exchange_rates['Time'] = pd.to_datetime(exchange_rates['Time'])
exchange_rates.sort_values('Time', inplace=True)
exchange_rates.reset_index(drop=True, inplace=True)
exchange_rates.head()
```

```python
#isolate Time and US_dollar columns
euro_to_dollar = exchange_rates[["Time", "US_dollar"]].copy()
euro_to_dollar.head() #shows the columns are isolated
```

```python
euro_to_dollar["US_dollar"].value_counts()    
```

```python
euro_to_dollar = euro_to_dollar[euro_to_dollar['US_dollar'] != '-']#drops rows where the - character appears
euro_to_dollar['US_dollar'] = euro_to_dollar['US_dollar'].astype(float) #convert to float type
euro_to_dollar.info()
```

# Rolling Mean
Now we will generate a line plot to visualise the evolution of the euro-dollar exchange rate

```python
import matplotlib.pyplot as plt
%matplotlib inline 
plt.plot(euro_to_dollar['Time'],
         euro_to_dollar['US_dollar'])
plt.show()
```

If we look at the line's shape, we see many small wiggles â€” rather than seeing a smooth line. The wiggles, however, have meaning: : they are the visual representation of the daily variation in the exchange rate. The rate goes up and down, up and down again, day to day. The rate only shows clear upward or downward trends in the longer run (months or years).

In this section we will be looking at the rolling mean for the `US_dollar` column using a moving window of 30 days

```python
euro_to_dollar['rolling_mean'] = euro_to_dollar['US_dollar'].rolling(window = 30).mean()
euro_to_dollar
```

# Coming up with an Idea
Our goal is to create a storytelling data visualization using the data we have. To create a story, we need to arrange our data into a series of events that show change.

I will be showing how the euro-dollar rate has changed during the coronavirus pandemic by showing the 2020 data and the 2016-2019 data as a baseline

# Coding the Graph

```python
financial_data = euro_to_dollar.copy(
                   )[(euro_to_dollar['Time'].dt.year >= 2015
                   ) & (euro_to_dollar['Time'].dt.year <= 2019)]
covid_pandemic = euro_to_dollar.copy(
                   )[(euro_to_dollar.Time.dt.year >= 2020
                   ) & (euro_to_dollar.Time.dt.year <= 2021)]
```

```python
### Adding the FiveThirtyEight style
import matplotlib.style as style
style.use('fivethirtyeight')

### Adding the plot
fig,ax = plt.subplots(figsize=(8,3))
ax.plot(financial_data['Time'],
        financial_data['rolling_mean'],
        linewidth=1, color='#A6D785')

### Highlighting the 2020-2021 period
ax.plot(covid_pandemic['Time'],
        covid_pandemic['rolling_mean'],
        linewidth=3, color='#e23d28')

### Highlihting the peak of the crisis
ax.axvspan(xmin=737799.0, xmax=737432.0, ymin=0.02,
           alpha=0.9, color='black')

### Adding separate tick labels
ax.set_xticklabels([])
ax.set_yticklabels([])

x = 735550.0
for year in ['2015', '2016', '2017', '2018', '2019', '2020', '2021']:
    ax.text(x, 1.03, year, alpha=0.5, fontsize=11)
    x += 365

y = 1.045
for rate in ['1.2', '1.3', '1.4', '1.5']:
    ax.text(735499.0, y, rate, alpha=0.5, fontsize=11)
    y += 0.067
    
## Adding a title and a subtitle    
ax.text(735550.0, 1.30, "Euro-USD rate during 2020-2021's covid pandemic",
        weight='bold')
ax.text(735550.0, 1.28, 'Euro-USD exchange rates between 2015 and 2021',
        size=12)

### Adding a signature
ax.text(735450.0, 1.01, 'Â©DATAQUEST' + ' '*94 + 'Source: European Central Bank',
        color = '#f0f0f0', backgroundcolor = '#4d4d4d',
        size=10)

### Add some transparency to the grid
ax.grid(alpha=0.5)

plt.show()
```
